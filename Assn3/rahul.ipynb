```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
import sklearn as skt
import scipy.stats as st
```

```python
img = cv2.imread("D:/Y/0.png",0)
```

```python
#bg=st.mode([img[0,0],img[0,-1],img[-1,0],img[-1,-1]])
bg=img[0,0]
```

```python
set1=np.array([],int)
for i in range(200):
    if np.sum(np.array(img[:,i]!=bg))>50:
        set1=np.append(set1,i)
        #print("Hello world")
pos1=int(np.mean(set1))

set2=np.array([],int)
for i in range(pos1+50,350):
    if np.sum(np.array(img[:,i]!=bg))>50:
        set2=np.append(set2,i)
        #print("Hello world")
pos2=int(np.mean(set2))

set3=np.array([],int)
for i in range(pos2+50,500):
    if np.sum(np.array(img[:,i]!=bg))>50:
        set3=np.append(set3,i)
        #print("Hello world")
pos3=int(np.mean(set3))
    


```

```python
images =[img[5:145,pos1-75:pos1+75], img[5:145,pos2-75:pos2+75], img[5:145,pos3-75:pos3+75] ]
for i in range(3):
    ax = plt.subplot(1, 3, i+1)
    plt.imshow(images[i].astype("uint8"), cmap='gray')
    #plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
ker=np.ones((5,4))#manipulate
```

```python
arr=[]
for i in images:
    ret,thresh = cv2.threshold(i.astype("uint8"), 0, 255, cv2.THRESH_BINARY_INV|cv2.THRESH_OTSU)
    img2=cv2.erode(thresh,ker,iterations=1)
    arr.append(img2)
```

```python
for i in range(3):
    ax = plt.subplot(1, 3, i+1)
    plt.imshow(arr[i].astype("uint8"), cmap='gray')
    #plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
for i in range(3):
    ax = plt.subplot(1, 3, i+1)
    plt.imshow(255-arr[i].astype("uint8"), cmap='gray')
    #plt.title(class_names[labels[i]])
    plt.axis("off")
```
